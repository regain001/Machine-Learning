# Functional API

```python
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,.8,1]) # [left,bottom,width,height] of rectangle.
axes.plot(x,y,'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title for the plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) # [left,bottom,width,height] of rectangle.
axes2 = fig.add_axes([0.1,0.4,.5,.5]) # sub axes!!
axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title for the plot')

axes2.plot(y,x,'b--')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title for the sub-plot')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b--')
axes[0].set_xlabel('x1')
axes[0].set_ylabel('y1')
axes[0].set_title('Title 1')

axes[1].plot(y,x,'r--')
axes[1].set_xlabel('x2')
axes[1].set_ylabel('y2')
axes[1].set_title('Title 2')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')
    
#separate(give some space between) two figure
fig.tight_layout()
```

```python
fig, axes = plt.subplots(figsize = (2, 2)) # u can change the figure size as ur wish!
axes.plot(x,y)
```

```python
# dpi = dot per inch like pixel per inch for using resolution in TV
fig, axes = plt.subplots(figsize = (2, 2), dpi=100) 

axes.plot(y,x,'r--')
```

```python
# jpg has low resolution & png has high resolution
fig.savefig('fig.png', dpi = 300)
```

```python
# Give/Track Two different title in one graph
fig, axes = plt.subplots(figsize = (4, 3), dpi=100) 

axes.plot(x,x**2,'r--', label = 'x**2')
axes.plot(x,x**3,'g--', label = 'x**3')
axes.legend()
```

```python
fig, axes = plt.subplots(figsize = (4, 3), dpi=100) 

axes.plot(x,x**2,color='blue', label = 'x**2',linewidth=3)
axes.plot(x,x**3,'g--', label = 'x**3',lw=4)
axes.legend()
```

```python
 fig, axes = plt.subplots(figsize = (4, 3), dpi=100) 

axes.plot(x,x**2,color='blue', label = 'x**2',linestyle=':',marker='>')
axes.plot(x,x**3,'g--', label = 'x**3',linestyle=':',marker='>')
axes.legend()
```

```python
fig, axes = plt.subplots(figsize = (6, 6), dpi=100) 

axes.plot(x,x**2,color='blue', label = 'x**2',linestyle=':',marker='o',markersize=8,markerfacecolor='blue',markeredgecolor='red')
axes.plot(x,x**3,'g--', label = 'x**3',linestyle='--',marker='o',markersize=8,markerfacecolor='green',markeredgecolor='yellow')
axes.legend()
```

```python
fig, axes = plt.subplots(figsize = (4, 3), dpi=100) 

axes.plot(x,x**2,color='blue', label = 'x**2')
axes.plot(x,x**3,'g--', label = 'x**3')
axes.legend()

# it's actually used for zooming ... 
axes.set_xlim(0,6) 
axes.set_ylim(0, 250)
```
