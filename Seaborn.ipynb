```python
import seaborn as sns
```

```python
# without calling plt.show(), it can give u  output
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
tips.tail()
```

```python
sns.distplot(tips['tip']) # distribution plot
```

```python
# tip = $2 have 40 times in DataSet..
# means many of the people just go for tip by $2
sns.distplot(tips['tip'],  kde=False, bins=40) # distribution plot
```

```python
# This small line (rug=True) means 
# At this position (those line), we have values for this perticular column of our DataSet. 
# For x = 11, there has no value for 'tip' column.For that, we can't see any tiny/small line at the
# point of 11 in X-axis.

# The more lines you have here, The more data you have there. (only for X-axis)
sns.distplot(tips['tip'], kde=True, hist=False, rug=True, bins=40) # distribution plot
```

```python
sns.distplot(tips['tip'], kde=False, hist=True, rug=True, bins=40) # distribution plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) 
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') # kind = scatter,hex..
```

```python
# The darker of your color is, The more values are located at that area
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # regression
```

```python
# all the numarical data column shows by pairplot()
# you can find the correlation between two continuous variable.And The ralation in between 0 to 1.
# we see here is positive correlation between total_bill and tip
sns.pairplot(tips)
```

```python
# relationship between total_bill, tip, size with column 'sex'
# Insights of these plot is, Male are gone for higher tip but female not
sns.pairplot(tips, hue='sex')
```

```python
# people who smoke or not
sns.pairplot(tips, hue='smoker')
```
